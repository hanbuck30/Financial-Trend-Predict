```python
fd = fd.iloc[:]

for column in fd.columns:
    if fd[column].dtype == 'object' and column not in ['TRD_DD', 'ISU_NM']:
        try:
            fd[column] = fd[column].str.replace(',', '').astype('float')
        except ValueError as e:
            print(f" {column}: {e}")
    elif column not in ['TRD_DD', 'ISU_NM']:
        fd[column] = fd[column].astype('float')

fd
```
